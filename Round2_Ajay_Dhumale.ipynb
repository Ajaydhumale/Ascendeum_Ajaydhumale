```python
df_1= df.groupby('bidder')[['prebid_win_count','win_count']].count().reset_index()
```

```python
df_1['total_wins'] = df_1['prebid_win_count']+df_1['win_count']
```

```python
df_1
```

```python
df_1['winning_rate'] = (df_1['total_wins']/total_bid)*100
```

```python
sorted_high_winning_rate =df_1.sort_values(by='winning_rate', ascending = False, ignore_index=True)
```
